```python
import pandas as pd
train_df = pd.read_csv("fraudTrain.csv")
test_df = pd.read_csv("fraudTest.csv")
```

```python
drop_cols = [
    'Unnamed: 0',
    'trans_date_trans_time',
    'cc_num']
train_df = train_df.drop(columns=drop_cols)
test_df = test_df.drop(columns=drop_cols)
```

```python
drop_cols = [
    'first', 'last',
    'street', 'city', 'state', 'zip',
    'dob',
    'trans_num'
]

train_df = train_df.drop(columns=drop_cols)
test_df = test_df.drop(columns=drop_cols)
```

```python
train_df = pd.get_dummies(train_df, drop_first=True)
```

```python
test_df = pd.get_dummies(test_df, drop_first=True)
```

```python
train_df
```

```python
test_df
```

```python
train_df, test_df = train_df.align(test_df, join='left', axis=1, fill_value=0)
```

```python
train_df = train_df.sample(10000, random_state=42)
test_df = test_df.sample(80000, random_state=42) #used this bex my dataset was having a large data
```

```python
X_train = train_df.drop('is_fraud', axis=1)
y_train = train_df['is_fraud']
```

```python
X_test = test_df.drop('is_fraud', axis=1)
y_test = test_df['is_fraud']
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000, class_weight='balanced')
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_lr = lr.predict(X_test)
print("accuracy_score",accuracy_score(y_test,y_pred_lr))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_lr))
print(confusion_matrix(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(class_weight='balanced', random_state=42)
```

```python
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_rf = rf.predict(X_test)
print("accuracy_score",accuracy_score(y_test,y_pred_rf))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
```
